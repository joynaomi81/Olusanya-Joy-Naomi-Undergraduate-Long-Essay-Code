Non-Tone-Marked Performance Results/Tone-Marked Performance Results

```python
import matplotlib.pyplot as plt
import numpy as np

# Models
models = ['Whisper-small', 'MMS-1B-all', 'AstralZander/yoruba_ASR']

# Tone-Marked Dataset (WER and TER in percentage)
wer_tone_marked = [66, 85, 77]
ter_tone_marked = [32, 52, 31]

# Non-Tone-Marked Dataset (WER and TER in percentage)
wer_non_tone_marked = [38, 50, 48]
ter_non_tone_marked = [4, 5, 6]

x = np.arange(len(models))  # label locations
width = 0.35  # width of bars

fig, ax = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

# Tone-Marked Plot
ax[0].bar(x - width/2, wer_tone_marked, width, label='WER', color='#1f77b4')
ax[0].bar(x + width/2, ter_tone_marked, width, label='TER', color='#ff7f0e')
ax[0].set_title('Tone-Marked Performance Results')
ax[0].set_ylabel('Error Rate (%)')
ax[0].set_xticks(x)
ax[0].set_xticklabels(models, rotation=10)
ax[0].legend()
ax[0].grid(axis='y', linestyle='--', alpha=0.7)

# Non-Tone-Marked Plot
ax[1].bar(x - width/2, wer_non_tone_marked, width, label='WER', color='#1f77b4')
ax[1].bar(x + width/2, ter_non_tone_marked, width, label='TER', color='#ff7f0e')
ax[1].set_title('Non-Tone-Marked Performance Results')
ax[1].set_xticks(x)
ax[1].set_xticklabels(models, rotation=10)
ax[1].legend()
ax[1].grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
